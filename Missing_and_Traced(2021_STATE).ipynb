```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('NCRB-2021_Table_15.2.csv')
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df.columns
```

```python
past_missing=df[['State/UT','Unrecovered / Untraced Missing Persons from Previous Years - Male','Unrecovered / Untraced Missing Persons from Previous Years - Female','Unrecovered / Untraced Missing Persons from Previous Years - Transgender']]
```

```python
#Unrecovered / Untraced Missing Persons from Previous Years
past_missing
```

```python
past_missing.rename(columns = {'Unrecovered / Untraced Missing Persons from Previous Years - Male':'Male',
                     'Unrecovered / Untraced Missing Persons from Previous Years - Female':'Female',
                     'Unrecovered / Untraced Missing Persons from Previous Years - Transgender':'Transgender'},
          inplace = True)
```

```python
past_missing.head()
```

```python
past_missing
```

```python
total_stat=past_missing[(past_missing['State/UT']=='Total State (S)')|(past_missing['State/UT']=='Total UT (S)') |(past_missing['State/UT']=='Total All India')]
```

```python
total_stat
```

```python
past_missing.drop([28,37,38],axis=0,inplace=True)
```

```python
states1=past_missing.iloc[:28] 
ut1=past_missing.iloc[29:] 
```

```python
states1
```

```python
ut1
```

```python
ax=states1.plot(x='State/UT', y=['Male','Female','Transgender'], kind="bar",figsize=(12,5))
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('No of missing Person')
plt.grid(axis='y')
plt.title('Unrecovered / Untraced Missing Persons from before 2021 in Indian states')
plt.tight_layout()
plt.savefig("Untraced Missing Persons from before 2021 in Indian states.png", format="png", dpi=1200)
```

```python
ax=ut1.plot(x='State/UT', y=['Male','Female','Transgender'], kind="bar",figsize=(12,5))
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('No of missing Person')
plt.grid(axis='y')
plt.title('Unrecovered / Untraced Missing Persons from before 2021 in Indian Union Territories')
plt.tight_layout()
```

```python
ax=sns.barplot(x = 'State/UT',
            y = 'Female',
            data = states1,
            estimator = np.median,
            errorbar=('ci', 0))

sns.set(rc={'figure.figsize':(15,6)})
# Show the plot
plt.xticks(rotation=90)
plt.title('Unrecovered / Untraced Missing Females from before 2021 in Indian States')
plt.ylabel('No. of females missing')
plt.xlabel('State')
plt.tight_layout()
plt.show()
```

```python
ax=sns.barplot(x = 'State/UT',
            y = 'Female',
            data = ut1,
            estimator = np.median,
            errorbar=('ci', 0))

sns.set(rc={'figure.figsize':(15,6)})
# Show the plot
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+.25, p.get_height()+0.01))
plt.xticks(rotation=90)
plt.title('Unrecovered / Untraced Missing Females from before 2021 in Indian union territories')
plt.ylabel('No. of females missing')
plt.xlabel('union territories')
plt.tight_layout()
plt.show()
```

```python
ax=sns.barplot(x = 'State/UT',
            y = 'Transgender',
            data = states1,
            estimator = np.median,
            errorbar=('ci', 0))

sns.set(rc={'figure.figsize':(15,6)})
# Show the plot
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()-0.25, p.get_height()+0.01))
plt.xticks(rotation=90)
plt.title('Unrecovered / Untraced Missing Transgender from before 2021 in Indian States')
plt.ylabel('No. of Transgender missing')
plt.xlabel('State')
plt.tight_layout()
plt.show()
```

```python
ax=sns.barplot(x = 'State/UT',
            y = 'Transgender',
            data = ut1,
            estimator = np.median,
            errorbar=('ci', 0))

sns.set(rc={'figure.figsize':(15,6)})
# Show the plot
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()-0.25, p.get_height()+0.01))
plt.xticks(rotation=90)
plt.title('Unrecovered / Untraced Missing Transgenders from before 2021 in Indian Union Territories')
plt.ylabel('No. of Transgender missing')
plt.xlabel('Union Territories')
plt.tight_layout()
plt.show()
```
